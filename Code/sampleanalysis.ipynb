```python
import plotly.figure_factory as ff
import numpy as np

# Remove rows with NaN values
df_config = df_config.dropna()

# Replace infinite values with NaN
df_config = df_config.replace([np.inf, -np.inf], np.nan)

# Again remove rows with NaN values if any were introduced in the previous step
df_config = df_config.dropna()

x1 = df_config['1 System']
x2 = df_config['2 System'] + 2
x3 = df_config['3 System'] + 4

hist_data = [x1, x2, x3]

group_labels = ['1 System', '2 System', '3 System']

colors = ['#835AF1', '#7FA6EE', '#B8F7D4']

# Create distplot with curve_type set to 'normal'
fig = ff.create_distplot(hist_data, group_labels, colors=colors, bin_size=.25, show_curve=False)

plt.tight_layout()
# Add title
fig.update_layout(title_text='Curve and Rug Plot')
fig.show()
```
